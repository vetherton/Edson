```python
import requests
from bs4 import BeautifulSoup

URL = "https://fbref.com/en/players/45a03961/nat_tm/Benjamin-Pavard-National-Team-Stats"
r = requests.get(URL)

soup = BeautifulSoup(r.content)
print(soup.prettify())
```

```python
roster = ['Lucas Martínez Quarta',
'Nicolás Tagliafico',
'Gonzalo Montiel',
'Germán Pezzella',
'Marcos Acuña',
'Cristian Romero',
'Nicolás Otamendi',
'Lisandro Martínez',
'Nahuel Molina',
'Leandro Paredes',
'Rodrigo De Paul',
'Ángel Di María',
'Exequiel Palacios',
'Giovani Lo Celso',
'Guido Rodríguez',
'Alexis Mac Allister',
'Valentin Carboni',
'Enzo Fernández',
'Julián Álvarez',
'Lionel Messi',
'Nicolás González',
'Alejandro Garnacho',
'Lautaro Martínez']

country_name = 'Argentina'
abbrv = 'ARG'
```

```python
import pandas as pd

country_df = pd.DataFrame(roster, columns=['Player'])
country_df
```

```python
import re

def extract_code_and_name(html_link):
    # Regular expression pattern to extract the code and name
    pattern = r'en/players/([a-zA-Z0-9]+)/([a-zA-Z-]+)'
    
    # Using re.search() to find the match
    match = re.search(pattern, html_link)
    
    if match:
        code = match.group(1)
        name = match.group(2)
        return code, name
    else:
        return None, None

URL = f"https://fbref.com/en/country/players/{abbrv}/{country_name}-Football-Players"
r = requests.get(URL)

soup = BeautifulSoup(r.content)
divs = soup.find_all('div', class_='section_content')

div = divs[0]

p_elements = div.find_all('p')

links = []
for p in p_elements:
    a_tag = p.find('a')
    if a_tag and 'href' in a_tag.attrs:
        link = a_tag['href']
        code, name = extract_code_and_name(link)
        text = a_tag.get_text(strip=True)
        links.append((text, code, name))

data_df = pd.DataFrame(links, columns=['Player', 'Code', 'Formatted'])

df = pd.merge(country_df, data_df, on='Player', how='left')

df['Link'] = df.apply(lambda row: f'https://fbref.com/en/players/{row["Code"]}/nat_tm/{row["Formatted"]}', axis=1)

df
```

```python
import time

all_data = []

for index, link in enumerate(df['Link']):
    time.sleep(5)
    r = requests.get(link)
    r.raise_for_status()
    
    soup = BeautifulSoup(r.content, 'html.parser')

    # print(soup.prettify())
    div = soup.find('div', id='div_stats_standard_nat_tm')

    if not div:
        print(f"No table found for link {link}, skipping...")
        continue

    tfoot = div.find('tfoot')

    if not tfoot:
        print(f"No tfoot found for link {link}, skipping...")
        continue

    rows = tfoot.find_all('tr')
    data = []
    first_row = True
    
    for row in rows:
        cols = row.find_all('td')
        cols = [col.get_text() for col in cols]
        header = row.find('th')
        header = header.get_text()
        
        cols.insert(0, header)
        if first_row:
            del cols[1:4]
            first_row = False

        else:
            del cols[1:3]

        cols.pop(-1)
        data.append(cols)

    thead = div.find('thead')
    rows = thead.find_all('tr')
    headers = rows[1]
    headers = headers.find_all('th')
    headers = [header.get_text() for header in headers]
    del headers[0:3]
    del headers[1]
    del headers[-1]

    player_name = df.loc[index, 'Player']
    
    for row in data:
        row.insert(0, player_name)

    headers.insert(0, 'Player Name')

    all_data.extend(data)
    
    # Convert to DataFrame and append to dfs
    df_table = pd.DataFrame(data)
    df_table.columns = headers
    
    print(df_table)

final_df = pd.DataFrame(all_data, columns=headers)
print(final_df)
```

```python
import numpy as np

final_df["Min"] = final_df["Min"].apply(lambda x: x.replace(',', ''))
final_df.replace('', np.nan, inplace=True)
final_df.fillna(0, inplace=True)

for col in final_df.columns:
    
    if (col != "Player Name" and col != "Comp"):
        final_df[col] = final_df[col].astype('float')
```
